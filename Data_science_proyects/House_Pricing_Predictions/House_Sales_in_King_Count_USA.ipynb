```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]    
X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)


print("number of test samples:", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

```python
from sklearn.linear_model import Ridge
```

```python
RigeModel = Ridge(alpha=0.1) 
RigeModel.fit(x_train, y_train)
test_score = RigeModel.score(x_test, y_test)
train_score = RigeModel.score(x_train, y_train)
print("R^2 test data: ",test_score)
```

```python
pr = PolynomialFeatures(degree=2)
x_train_pr1 = pr.fit_transform(x_train)
x_test_pr1 = pr.fit_transform(x_test) 
RigeModel = Ridge(alpha=0.1) 
RigeModel.fit(x_train_pr1, y_train)
test_score1 = RigeModel.score(x_test_pr1, y_test)
train_score1 = RigeModel.score(x_train_pr1, y_train)
print("R^2 test data: ",test_score1,"R^2 train data: ",train_score1)
```
